```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import math
```

```python
def graph_3d(graph_num, x, y, z, heading, x_label, y_label, z_label, line):
    fig = plt.figure(graph_num)
    ax = fig.add_subplot(111, projection="3d")
    ax.plot(x, y, z ,line) 
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label) 
    ax.set_zlabel(z_label) 
    ax.set_title(heading)
    
def surface_graph(graph_num, min_x, max_x, x_stride, min_y, max_y, y_stride, heading, x_label, y_label, z_label, color, line_width=0, antia=True):
    fig = plt.figure(graph_num)
    ax = fig.add_subplot(111, projection="3d")
    x = np.arange(min_x, max_x, x_stride)
    y = np.arange(min_y, max_y, y_stride)
    X, Y = np.meshgrid(x, y)
    Z = np.exp(-(X**2 + Y**2)/2.0) #!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label) 
    ax.set_zlabel(z_label) 
    ax.set_title(heading)
    ax.plot_surface(X,Y,Z,linewidth=line_width, antialiased=antia,color=color)
    
def wire_graph(graph_num, min_x, max_x, x_stride, min_y, max_y, y_stride, heading, x_label, y_label, z_label, color, r_stride=10, c_stride=10):
    fig = plt.figure(graph_num)
    ax = fig.add_subplot(111, projection="3d")
    x = np.arange(min_x, max_x, x_stride)
    y = np.arange(min_y, max_y, y_stride)
    X, Y = np.meshgrid(x, y)
    Z = np.exp(-(X**2 + Y**2)/2.0) #!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    ax.plot_wireframe(X,Y,Z,rstride=r_stride, cstride=c_stride)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label) 
    ax.set_zlabel(z_label) 
    ax.set_title(heading)
```

```python
t = []
xplot = []
yplot = []
#
# define the parametric equations for a helix
#
for n in range(100) :
    t.append(.1*n*math.pi)
    xplot.append(math.cos(t[n]))
    yplot.append(math.sin(t[n]))
    
graph_3d(1, xplot, yplot, t, "A Helix", "Cos(t)", "Sin(t)", "t", "r*")
surface_graph(2, -5, 5, 0.1, -5,5,0.1,"Graph", "X", "Y", "Z", "aquamarine")
wire_graph(3,-5, 5, 0.1, -5,5,0.1,"Graph", "X", "Y", "Z", "aquamarine")
```
